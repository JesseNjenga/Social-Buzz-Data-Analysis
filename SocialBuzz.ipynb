**Top 5 Categories with the Largest Popularity**



```python
#import pandas
import pandas as pd
```

```python
#load datasets
data = pd.read_csv('Content.csv')
data1 = pd.read_csv('Reactions.csv')
data2 = pd.read_csv('ReactionTypes.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
#unique elements in Category and Content_type
print(data['Category'].unique())
print('\n', data['Type'].unique())
```

```python
def clean_data(data):
  # Replace column names spaces with _
  data.columns = data.columns.str.replace(' ', '_')

  # Drop unwanted columns
  data.drop(['Unnamed:_0', 'User_ID', 'URL'], axis=1, inplace=True)

  # Rename Type column to Content_type
  data.rename(columns={'Type': 'Content_type'}, inplace=True)

  #remove "" in elements of Category
  data['Category'] = data['Category'].str.replace('"', '')



  return data

content = clean_data(data)
```

```python
#print unique values for Type
print(data1['Type'].unique())
```

```python
def clean_data1(data):
  # Replace column names spaces with _
  data.columns = data.columns.str.replace(' ', '_')

  # Drop unwanted columns
  data.drop(['Unnamed:_0', 'User_ID'], axis=1, inplace=True)

  # Rename Type column to Content_type
  data.rename(columns={'Type': 'Reaction_type'}, inplace=True)

  #drop null rows
  data.dropna(inplace=True)


  return data

reaction_types = clean_data1(data1)
```

```python
reaction_types.info()
```

```python
reaction_types.head()
```

```python
data2.info()
```

```python
data2.head()
```

```python
def clean_data2(data):

  # Drop unwanted columns
  data.drop(['Unnamed: 0',], axis=1, inplace=True)

  # Rename Type column to Content_type
  data.rename(columns={'Type': 'Reaction_type'}, inplace=True)


  return data

reactions = clean_data2(data2)
```

```python
reactions.info()
```

```python
#merge the 3 datasets content & reaction_types on  Content_ID and reactions on Reaction_type
df = pd.merge(content, reaction_types, on='Content_ID')
df = pd.merge(df, reactions, on='Reaction_type')
```

```python
df.info()
```

```python
#random
df.sample(10)
```

```python
#5 categories with the largest popurality
df.groupby('Category')['Score'].sum().sort_values(ascending=False).head(5)
```

```python
#export df CSV file
df.to_csv('social_buzz.csv')
```

```python
# Check for missing values
print(df.isnull().sum())

# Ensure all categories are strings
df['Category'] = df['Category'].astype(str)

# Perform the groupby operation again
top_categories = df.groupby('Category')['Score'].sum().sort_values(ascending=False).head(5)
print(top_categories)
```
